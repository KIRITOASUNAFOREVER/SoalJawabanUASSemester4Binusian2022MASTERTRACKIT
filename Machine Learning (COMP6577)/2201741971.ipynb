```
#NOMOR 2
#[15%] Lakukan logistic regression untuk memprediksi harga (price) dari masing-masing listing!
#Jelaskan bagaimana feature engineering dan feature selection mempengaruhi hasil dariprediksi Anda! 
#Apa metrik yang Anda gunakan untuk menghitung akurasi dari prediksi harga?


# Import Library yang dibutuhkan
import pandas as pd
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import r2_score

#Membuat data frame dari File CSV menggunakan pandas.
#Karena Sesuai Ketentuan Soal, hanya 12 Fitur yang dipilih, saya Menggunakan usecols
df = pd.read_csv("dataset.csv", usecols = ['accommodates','bathrooms','bedrooms','beds','guests_included','minimum_nights','maximum_nights','reviews_per_month','review_scores_rating' , 'number_of_reviews','number_of_reviews_ltm','price'])

#FETURE ENGINEERING
# Saya memilih 12 Fitur tersebut karena saling berkaitan satu sama lain, dimana untuk memprediksi harga, perlu dipertikan dari akomodasinya,
# minimum menginap berapa hari dan maksimalnya, review per bulan, jumlah review, jumlah lamar mandi, kamar tidur, ranjang
# Jadi, Karena hal itulah saya memilih 12 Fitur tersbut karena sangat berkaitan.

#Menghapus Baris Jika terdapat data NaN (Null Value)
df = df.dropna(axis='rows')

#Karena Fitur Price dalam bentuk String, Saya dalam hal ini Ubah ke Float
df['price'] = df.price.str.extract('(\d)').astype(float)

#Ketika Saya ingin Fitting ke Logistic Regression, terjadi Error "Cannot Convert String to Float"
#Maka Dari itu saya Meng-DROP Baris(ROW) jika ada yang STRING, saya lakukan hal ini ke semua Fitur sebagai berikut:
#1) UNTUK FITUR ACCOMMODATES
for count in df.index:
    try:
      float(df['accommodates'][count])
    except:
      df = df.drop(count)
df.astype({'accommodates' : 'float'})

#2) UNTUK FITUR BATHROOMS
for count in df.index:
    try:
      float(df['bathrooms'][count])
    except:
      df = df.drop(count)
df.astype({'bathrooms' : 'float'})

#3) UNTUK FITUR BEDROOMS
for count in df.index:
    try:
      float(df['bedrooms'][count])
    except:
      df = df.drop(count)
df.astype({'bedrooms' : 'float'})

#4) UNTUK FITUR BEDS
for count in df.index:
    try:
      float(df['beds'][count])
    except:
      df = df.drop(count)
df.astype({'beds' : 'float'})

#5) UNTUK FITUR GUEST_INCLUDED
for count in df.index:
    try:
      float(df['guests_included'][count])
    except:
      df = df.drop(count)
df.astype({'guests_included' : 'float'})

#6) UNTUK FITUR MINIMUM_NIGHTS
for count in df.index:
    try:
      float(df['minimum_nights'][count])
    except:
      df = df.drop(count)
df.astype({'minimum_nights' : 'float'})

#7) UNTUK FITUR MAXIMUM_NIGHTS
for count in df.index:
    try:
      float(df['maximum_nights'][count])
    except:
      df = df.drop(count)
df.astype({'maximum_nights' : 'float'})

#8) UNTUK FITUR REVIEWS_PER_MONTH
for count in df.index:
    try:
      float(df['reviews_per_month'][count])
    except:
      df = df.drop(count)
df.astype({'reviews_per_month' : 'float'})

#9) UNTUK FITUR REVIEW_SCORES_RATING
for count in df.index:
    try:
      float(df['review_scores_rating'][count])
    except:
      df = df.drop(count)
df.astype({'review_scores_rating' : 'float'})

#10) UNTUK FITUR NUMBER_OF_REVIEWS
for count in df.index:
    try:
      float(df['number_of_reviews'][count])
    except:
      df = df.drop(count)
df.astype({'number_of_reviews' : 'float'})

#11) UNTUK FITUR NUMBER_OF_REVIEWS_LTM
for count in df.index:
    try:
      float(df['number_of_reviews_ltm'][count])
    except:
      df = df.drop(count)
df.astype({'number_of_reviews_ltm' : 'float'})

#FEATURE SELECTION
#Memasukkan Fitur-Fitur ke Variabel X sebagai Variabel INDEPENDEN.
x = df[['accommodates','bathrooms','bedrooms','beds','guests_included','minimum_nights','maximum_nights','reviews_per_month','review_scores_rating' , 'number_of_reviews','number_of_reviews_ltm']]
#Memasukkan Fitur ke Variavel Y sebagai Variabel DEPENDEN yang akan DIPREDIKSI
y = df['price']

#Membagi Dataset Menajdi 80% untuk training dan 20 % untuk testing ditunjukkan dari test size=0.2.
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)

#Kita Lakukan Scalling Data Agar Lebih Teratur dan Rapi
x_train = preprocessing.scale(x_train)
x_test = preprocessing.scale(x_test)

#Sesuai Soal, gunakan LOGISTIC REGRESSION, disini saya menggunakan max_iter karena dataset setelab operasi diatas masih sangat besar
print("LOGISTIC REGRESSION UNTUK MEMPREDIKSI PRICE\n")
model = LogisticRegression(max_iter=100000)

#Fitting Nodel Logistic Regression saya dengan variabel training dari Variabel Independen dan Variabel Dependen
model.fit(x_train,y_train)

#Memprediksi visibilitas model terlatih pada Data Training dan Data Testing.
y_train_pred = model.predict(x_train)
y_test_pred = model.predict(x_test)

#Mencetak Nilai Prediksi dan Nilai Exact untuk melihat perbedaanya.
print("Mencetak Nilai Prediksi",y_train_pred)
print("Mencetak Nilai Exact",y_train)

#UNTUK METRIK PREDIKSI HARGA, Saya Menggunakan Metrik R2 SCORE baik di dalam Train Set dan Test Set
#Dari situ kita dapat Menentukkan Akurasi Prediksi kita.
#Mencetak r2 Score
print("r2 Score di dalam Train set adalah:\t{:0.3f}".format(-r2_score(y_train,y_train_pred)))
print("r2 Score di dalam Test set adalah :\t{:0.3f}".format(-r2_score(y_test,y_test_pred)))

#Mencetak Akurasi di dalam Train Set dan di dalam Test Set.
print("Akurasi di dalam Train set adalah:\t{:0.3f}%".format(-r2_score(y_train,y_train_pred)*100))
print("Akurasi di dalam Test set adalah :\t{:0.3f}%".format(-r2_score(y_test,y_test_pred)*100))
```
